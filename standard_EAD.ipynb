automated EDAs
1. summary tools - The package summarizes the data frame columns, including the statistics, frequency, distribution, and missing data
2. SweetViz - Python package for the automatic EDA process where the package aims explicitly to compare datasets and visualize the target label. Note, it can perform EDA without any data label.
3. D-Tale - to view and analyze our Pandas DataFrame structure
4. y-data profiling - a package that offers a nice quick report of the dataset
